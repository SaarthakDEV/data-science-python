```python
for i in range(8, 51):
    a = requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    if(a.status_code == 200):
        print(f"Page {i} downloaded successfully")
        with open(f"htmls/page{i}.html", "w", encoding="utf-8") as f:
            f.write(a.text)
    else:
        print(f"Error downloading Page {i}")
```

```python
from bs4 import BeautifulSoup
```

```python
with open("htmls/page1.html") as f:
    content = f.read()
soup = BeautifulSoup(content, "html.parser")
```

```python
type(soup)
```

```python
h3s = soup.find_all("h3")
```

```python
h3s.text
```

```python
for h3 in h3s:
    print(h3.find("a")["title"])
```

```python
articles = soup.find_all("article.product_pod")
```

```python
articles
```

```python
articles = soup.select("article.product_pod")
```

```python
articles
```

```python
items = []
for article in articles:
    title = article.find("h3").find("a")["title"]
    price = article.select_one("p.price_color").text
    rating = article.select_one("p.star-rating")["class"][1]
    items.append({
        "title": title,
        "price": price,
        "rating": rating
    })

#for p in price:
#    print(p.select("p.price_color"))
```

```python
items
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(items)
```
