```python
import matplotlib.pyplot as plt
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
pivot_table = flights.pivot(index="month", columns="year", values="passengers")
 
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
plt.show()
```

#### Using the fmri dataset from Seaborn (sns.load_dataset("fmri")), create a line plot showing signal over timepoint, separated by region.

Add markers to the lines.

Use a legend to distinguish regions.

```python
fmri = sns.load_dataset("fmri")
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri, hue="region", marker='o', markersize=8, markerfacecolor='red', markeredgecolor='blue')
plt.legend()
```
