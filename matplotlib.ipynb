```python
import numpy as np
years = [2016,2017,2018,2019,2020,2021,2022,2023,2024,2025]
pyt.plot(years, np.random.rand(10), color="red", linestyle="--", label="Kohli")
pyt.plot(years, np.random.rand(10), color="blue", linestyle="-.", label="Rohit")
pyt.plot(years, np.random.rand(10), color="green",linestyle=":", label="Sehwag")
pyt.style.use("ggplot")
pyt.legend()
pyt.xlabel("Years")
pyt.ylabel("Runs")
pyt.grid(True)
pyt.tight_layout()
```

```python
with pyt.xkcd():
    pyt.plot(years, kohli, label="Kohli")
    pyt.plot(years, sehwag, label="Sehwag")
    pyt.title("Epic Battle of the Batsmen")
    pyt.legend()
    pyt.show()
```

# Bar charts 

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

pyt.bar(years, runs, width=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
 
x = np.arange(len(years))  # index positions
width = 0.25
plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")
 
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
plt.bar(x, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x, kohli, width=width, label="Kohli")
 
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
num_players = np.array([1,2,3,4])
runs_1yrs = [500, 0, 0, 300]
runs_2yrs = [700,200,0,600]
runs_3yrs = [1100, 900, 500, 600]
runs_4yrs = [1500, 1400, 800, 1100]
runs_5yrs = [1800, 1600, 1100, 900]
```

```python
plt.barh(runs_1yrs, players)
```

```python
width=0.20
```

```python
plt.barh(num_players - (2 * width),runs_1yrs,height=0.20, color="red")
plt.barh(num_players - width,runs_2yrs,height=0.20, color="blue")
plt.barh(num_players,runs_3yrs,height=0.20, color="green")
plt.barh(num_players + width,runs_4yrs,height=0.20, color="yellow")
plt.barh(num_players + (2 * width),runs_5yrs,height=0.20, color="black")
plt.yticks(num_players, players)
plt.legend()
```

```python
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]
 
plt.bar(players, runs, color="skyblue")

for i in range(len(players)):
    plt.text(i, runs[i] + 50, str(runs[i]), ha="center")
plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
```

# Pie charts

```python
import matplotlib.pyplot as plt
 
plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
 
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': '#fffff0', 'linewidth': 8, 'linestyle': ':'}
)
```

```python
explode = [0.1, 0, 0.4, 0]
plt.pie(runs, labels=labels, shadow=True, explode=explode)
```

# Stackplot in matplotlib

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]

colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']
plt.figure(figsize=(6,6))
plt.pie(time_spent, labels=activities, colors=colors, autopct='%1.1f%%', startangle=90)
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
 
labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']
 
plt.figure(figsize=(10,6))
```

```python
plt.stackplot(days, studying, playing, watching_tv, sleeping,
              labels=labels, colors=colors)
plt.legend(loc="best")
```
